```python
import pandas as pd
```

```python
texto = 'ScienceDirect_citations_1667405027128.txt'
```

```python
diccs = []
campos = ['Autores', 'Titulo', 'Fuente', '#Publicacion', 'Anio', 
          'Index','Cod ISSN','DOI URL', 'URL', 'Abstract', 'Keywords']

with open(texto, 'r') as f:
    buffer = []
    for l in f.readlines():
        if len(l) > 1:
            buffer.append(l)
        else:
            diccionario = dict(zip(campos,buffer))
            diccs.append(diccionario)
            buffer.clear()
```

```python
pd.DataFrame.from_dict(diccs)
```
